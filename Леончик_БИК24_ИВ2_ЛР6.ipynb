Написать программу, которая с помощью цикла создаёт список чисел от 1 до 10, а также списки их квадратов и кубов. В конце необходимо вывести три соответствующих списка.

```python
n = []
n2 = []
n3 = []
for i in range(1, 11):
  n.append(i)
  n2.append(i**2)
  n3.append(i**3)
print(n, n2, n3)
```

С клавиатуры ввоится список чисел до тех пор, пока не введён 0.

```python
l = []
i = 1
while i != 0:
  l.append(i)
print(min(l))
print(max(l))
print(sorted(l))
```
